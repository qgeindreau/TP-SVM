Attention, la base MNIST contient un nombre de points importants. Le temps de

```python
from sklearn.datasets import load_digits
import pandas as pd
import numpy as np
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score

digits=load_digits()

X=digits['data']
y=digits['target']

clf = SVC().fit(X,y)

score_clf = np.mean(
    cross_val_score(clf,
                    X,
                    y,
                    cv=5,
                    n_jobs=-1,
                    scoring="accuracy"))
print("Précision modèles de base",score_clf)
```

```python
from hpsklearn import svc, HyperoptEstimator, svr, any_preprocessing
from hyperopt import tpe, rand

#On garde la fonction de perte de base étant 1 -accuracy

estim = HyperoptEstimator(classifier=svc('MySvr'),
                          preprocessing=any_preprocessing('prepro'),
                          algo=rand.suggest,
                          max_evals=300,
                          trial_timeout=2,
                          n_jobs=-1)
estim.fit(X,y,
          n_folds=5,
          cv_shuffle=False,
          random_state=None)

print(estim.best_model())
```
